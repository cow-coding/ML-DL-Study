# 학습이 잘 되는 모뎅
- 사용할 레이어
  - tf.keras.layers.BatchNormalization()
  - tf.keras.layers.Activation('swish')
- 데이터
  - 보스턴 집값: 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv
  - 아이리스 품종: 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'

```
# library
import tensorflow as tf
import pandas as pd
```

```
# past data
url =  'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(url)

cause = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
effect = boston[['medv']]

print(cause.shape, effect.shape)
```

```
# model making

X = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)

model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
# model fitting
model.fit(cause, effect, epochs=1000)
```

```
print(model.predict(cause[0:5]))
print(effect[0:5])
```

# 아이리스 품종

```
url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(url)
iris = pd.get_dummies(iris)

cause = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
effect = iris[['품종_setosa', '품종_versicolor', '품종_virginica']]

print(cause.shape, effect.shape)
```

```
X = tf.keras.layers.Input(shape=[4])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
model.fit(cause, effect, epochs=1000, batch_size=150)
```

```
print(model.predict(cause[0:5]))
print(effect[0:5])
```
